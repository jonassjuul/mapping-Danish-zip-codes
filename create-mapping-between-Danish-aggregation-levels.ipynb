```python
# Read file containing mappings between zip codes, municipalities, and regions
df = pd.read_csv('input-files/Zip_municipality_region.csv',delimiter='\t',skiprows=4)

# Split codes and names into different columns for zip, munipality, and region
df[['zip-code','zip-name']] = df.postnumre.str.split(" ",1,expand=True)
df[['municipality-code','municipality-name']] = df.kommuner.str.split(" ",1,expand=True)
df[['region-code','region-name']] = df.regioner.str.split(" ",1,expand=True)

# Create column with province name
df['province-name'] = df['municipality-name'].map(dic_KommuneToLandsdel)
df['municipality-type'] = df['municipality-name'].map(dic_KommuneToType)

# Delete old zip, municipality, and regions
df = df.drop(['postnumre','kommuner','regioner'],axis=1)
```
